>
> Not all code requires an automated test suite. However, if you bother to write unit tests, write the tests first. Your first code after that should fail every test. Then the tests become your TODO list going forward. Keep rewriting the code and re-running the tests until the code passes every test.Thank you, Professor Huntley!
>
> 如何理解Lesson4，关于向前失败中关于测试的内容，帮忙讲解：
并非所有代码都需要自动化测试套件。但是，如果您费心编写单元测试，请先编写测试。之后的第一个代码应该在每个测试中都失败。然后测试就变成了你的待办事项列表。继续重写代码并重新运行测试，直到代码通过每个测试为止。谢谢亨特利教授！

Failing forward is a classic approach popularized by the Extreme Programming movement a couple decades ago. The general idea is that in order to truly know if you have "solved" a problem, define exactly _what a solution looks like_ beforehand. The tests are exactly that. Ways of validating that the programming problem has been solved. If the problem itself changes, then update the tests accordingly. 

Here is how the Agile Alliance explains it. 
> Test-First Programming
Instead of following the normal path of:
>
> develop code -> write tests -> run tests
>
> The practice of Test-First Programming follows the path of:
>
>Write failing automated test -> Run failing test -> develop code to make test pass -> run test -> repeat
>
>As with Continuous Integration, Test-First Programming reduces the feedback cycle for developers to identify and resolve issues, thereby decreasing the number of bugs that get introduced into production.

A "literal" is 'source" data whose type can be detected by looking at it. No interpreter. No instanceof() or type() calls. Just looking at it with your eyes. 

In other words, a literal is not a variable or a constant. 

- `2` is a literal
- `"2"` is a literal
- `True` is a literal

Meanwhile
- `x` is a variable
- `math.pi` is a constant

You can usually tell if something is a literal because it will appear in your code editor in green or red. The colors indicate that the editor detects the data type just by "looking at" the data. So, variables always apear in black because the data type is unknown without evaluating the variable to see its value. Similarly, the constant `pi` below is in blue. It is treated like a special kind of variable and is not a literal value.
